```python
import numpy as np
import matplotlib.pyplot as plt
from pylab import *
import statsmodels.api as sm
np.__version__
```

```python
nsample = 100
x = np.linspace(1,10,nsample)
X = sm.add_constant(x)
beta = np.array((1,10))
e = np.random.normal(size = nsample)
y = np.dot(X,beta)+e
#print(x.shape,beta.shape,y.shape)
#print(beta)
model = sm.OLS(y,X)
results =  model.fit()
results.params
results.summary()
y_fitted = results.fittedvalues

fig, ax = plt.subplots(figsize=(8,6))
ax.plot(x,y,'o',label='data')
ax.plot(x,y_fitted,'r--',label='OLS')
ax.legend(loc='best')
ax.axis((1, 2, 1, 25))
show()
```

```python
nsample = 100
x = np.linspace(1,10,nsample)
x2 = np.column_stack((x,x**2))
X = sm.add_constant(x2)
beta = np.array((1,0.1,10))
e = np.random.normal(size = nsample)
y = np.dot(X,beta)+e
#print(x.shape,beta.shape,y.shape)
#print(beta)
model = sm.OLS(y,X)
results =  model.fit()
results.params
results.summary()
y_fitted = results.fittedvalues

fig, ax = plt.subplots(figsize=(8,6))
ax.plot(x,y,'o',label='data')
ax.plot(x,y_fitted,'r--',label='OLS')
ax.legend(loc='best')
#ax.axis((1, 2, 1, 25))
show()
```

```python
nsample = 100
nsample = 50
group = np.zeros(50,int)
group[20:40]=1
group[40:]=2
dumped = sm.categorical(group,drop=True)
x = np.linspace(0,20,nsample)
x2 = np.column_stack((x,dumped))
X = sm.add_constant(x2)
beta = np.array((1,0.1,10))
beta = np.array((10,1,1,3,8))
e = np.random.normal(size = nsample)
y = np.dot(X,beta)+e
model = sm.OLS(y,X)
results =  model.fit()
results.params
results.summary()
y_fitted = results.fittedvalues

fig, ax = plt.subplots(figsize=(8,6))
ax.plot(x,y,'o',label='data')
ax.plot(x,y_fitted,label='OLS')
ax.legend(loc='best')
show()
```
